```python
import numpy as np
import pandas as pd
import networkx as nx
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/data_tr.csv')
df
```

```python
G = nx.Graph()

for _, row in df.iterrows():
    G.add_edge(row['u'], row['v'], t=row['t'], h=row['h'])
```

```python
def generate_recommendations(user_id):
    neighbors = list(G.neighbors(user_id))
    recommendations = []

    neighbor_scores = {}
    for neighbor in neighbors:
        edge_data = G[user_id][neighbor]
        t_score = edge_data['t'] / 99.0
        h_score = edge_data['h']
        neighbor_scores[neighbor] = t_score * h_score

    sorted_neighbors = sorted(neighbors, key=lambda x: neighbor_scores[x], reverse=True)

    recommendations = sorted_neighbors[:10]

    return recommendations
```

```python
with open(f'recommendations.txt', 'w') as f:
  for user_id in df['u'].unique():
    recommendations = generate_recommendations(user_id)
    f.write(f'{user_id}: {",".join(map(str, recommendations))}\n')
```
