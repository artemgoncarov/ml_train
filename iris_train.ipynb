```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.tree import plot_tree, DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import cross_val_score
```

```python
train_data = pd.read_csv('https://stepik.org/media/attachments/course/4852/train_iris.csv')
train_data
```

```python
test_data = pd.read_csv('https://stepik.org/media/attachments/course/4852/test_iris.csv')
test_data
```

```python
train_data.drop(['Unnamed: 0'], axis=1)
test_data.drop(['Unnamed: 0'], axis=1)
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=5)
```

```python
X_train = train_data.drop(['species'], axis=1)
y_train = train_data.species
```

```python
X_test = test_data.drop(['species'], axis=1)
y_test = test_data.species
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
scores_data = pd.DataFrame()
np.random.seed(0)
```

```python
for max_depth in range(1, 100):
  clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=max_depth)
  clf.fit(X_train, y_train)
  train_score = clf.score(X_train, y_train)
  test_score = clf.score(X_test, y_test)

  mean_cross_val_score = cross_val_score(clf, X_train, y_train, cv=5).mean()

  temp_score_data = pd.DataFrame({'max_depth': [max_depth], 'train_score': [train_score], 'test_score': [test_score], 'cross_val_score': [mean_cross_val_score]})
  scores_data = scores_data.append(temp_score_data)
```

```python
scores_data
```

```python
scores_data_long = pd.melt(scores_data, id_vars=['max_depth'], value_vars=['test_score', 'cross_val_score'], var_name='set_type', value_name='score')
scores_data_long
```

```python
sns.lineplot(x='max_depth', y='score', hue='set_type', data=scores_data_long)
```
